```python
def sigmoid(z):
    s = 1/(1+np.exp(-z))
    return s
```

 The model is z=wx+b

 Loss=-ylog(a)-(1-y)log(1-a)

```python
def forward_backward_pass(x_train,y_train,w,b):

    A=sigmoid(np.dot(w.transpose(),x_train)+b)
    cost=-1*np.sum(y_train*np.log(A)+(1-y_train)*np.log(1-A))
    #/num_train    
    dw=np.dot(x_train,(A-y_train).transpose())/num_train
    db=np.sum(A-y_train)/num_train 
    
    grads={'dw': dw, 'db': db}
    return grads, cost
```

```python
def predict(w,b,X):
    y=np.zeros((1,X.shape[1]))
    w=w.reshape(X.shape[0],1)
    
    A=sigmoid(np.dot(w.transpose(),X)+b)
    for i in range(A.shape[1]):
        if(A[0,i] > 0.5):
            y[0,i]=1
        else:
            y[0,i]=0
    return y
```

```python
def optimize(x_train,y_train,x_test,y_test,iterations,learning_rate):
    dim=x_train.shape[0]
    w=np.zeros((dim,1))
    b=0
    for i in range(iterations):
        print(i)
        grads,cost=forward_backward_pass(train_set_x,train_set_y_orig,w,b)
        w=w-learning_rate*grads['dw']
        b=b-learning_rate*grads['db']
        print("Cost",cost)
    Y_prediction_test = predict(w, b, x_test)
    Y_prediction_train = predict(w, b, x_train)

    print("train accuracy: {} %".format(100 - np.mean(np.abs(Y_prediction_train - y_train)) * 100))
    print("test accuracy: {} %".format(100 - np.mean(np.abs(Y_prediction_test - y_test)) * 100))

        
```

```python
optimize(train_set_x,train_set_y_orig,test_set_x,test_set_y_orig,1500,0.001)
```
